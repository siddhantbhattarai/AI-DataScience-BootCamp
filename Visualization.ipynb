```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/siddhantbhattarai/practice-datasets/main/data.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.plot()
```

```python
df.plot(kind='scatter', x = 'Duration', y = 'Calories')
```

```python
df.plot.scatter(x='Duration', y='Pulse')
```

```python
df.plot.scatter(x='Duration', y='Maxpulse')
```

```python
df.plot(kind='hist') 
# df.plot.hist()
```

```python
df.plot.area()
```

```python
# df.plot(kind='bar')
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh()
```

```python
df.plot.box()
```

```python
df.plot.hexbin(x='Maxpulse', y='Calories', gridsize=18)
```

```python
df.plot.kde()
```
