What is Pandas? 
- Pandas is a Python library used for working with data sets.
- It has functions for analyzing, cleaning, exploring, and manipulating data.
- The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

Why Use Pandas?
- Pandas allows us to analyze big data and make conclusions based on statistical theories.
- Pandas can clean messy data sets, and make them readable and relevant.
- Relevant data is very important in data science.

```python
# Working with titanic dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the csv file 
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
# View the first few rows
df.head()
```

```python
# View last few rows
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Checking the missing values
missing_values = df.isnull().sum()
print(f"Missing Values is:\n {missing_values}")
```

```python
print("Duplicated Values if present:", df.duplicated().sum())
```

```python
# Filling the age values with median age
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Filling missing cabin values with placeholder 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Filling missing Embarked values with most frequent values (mode)
most_frequent_values = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_frequent_values, inplace=True)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Age'])
plt.title('Age Distribution with Outliners')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution with Outliners')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

```python
age_bins = [0, 16, 32, 48, 64, 80]
age_labels = ['Child', 'Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
```

```python
df.head()
```

```python
fare_bins = [0, 7.91, 14.454, 31, float('inf')]
fare_labels = ['Low', 'Medium', 'High', 'Very High']
df['FareCategory'] = pd.cut(df['Fare'], bins=fare_bins, labels=fare_labels)
```

```python
df.head()
```

```python
# Drop the unnecessary columns
titanic_data = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
titanic_data.head()
```

```python
# Calculate the survival rate by passenger class
survival_rate_by_class = titanic_data.groupby('Pclass')['Survived'].mean().reset_index()
```

```python
# Create a bar plot
plt.figure(figsize=(8,6))
plt.bar(survival_rate_by_class['Pclass'], survival_rate_by_class['Survived'])
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.xticks(survival_rate_by_class['Pclass'], ['1st Class', '2nd Class', '3rd Class'])
plt.show()
```

```python
embarked_counts = titanic_data['Embarked'].value_counts()

# Create a pie chart
plt.figure(figsize=(8,6))
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Proportion of Passenger by Embarkation Port')
plt.show()
```
